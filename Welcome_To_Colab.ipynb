```

def final_grade(term_score, exam_score):
    grade = (0.4 * term_score) + (0.6 * exam_score)

    if grade > 60:
        return "Success"
    else:
        return "failure"

term_score = float(input("Note intermédiaire: "))
exam_score = float(input("Note finale: "))

print(final_grade(term_score, exam_score))
```

```
def check_passport(number):
    if len(number) == 6 and number.isdigit():
        if number.startswith("75") or number.startswith("76"):
            return "new"
        else:
            return "old"
    else:
        return "wrong number"

line = input("Numéro de passeport: ")

print(check_passport(line))
```

```
n = int(input("Taille de la matrice: "))
matrix = []

for i in range(n):
    row = []
    for j in range(n):
        if i == j:
            row.append(1)
        else:
            row.append(0)
    matrix.append(row)

#ci-dessous ce code pour vérification (ne pas effacer)
for i in range(n):
    print(matrix[i])
```

```
countries = {
    'Sweden': ['Stockholm', 'Goteborg', 'Malmo'],
    'Norway': ['Oslo', 'Bergen', 'Trondheim'],
    'England': ['London', 'Birmingham', 'Manchester'],
    'Germany': ['Berlin', 'Hamburg', 'Munich'],
    'France': ['Paris', 'Marseille', 'Toulouse']
}

city = input("Entrez le nom d'une ville: ")

found = False
for country in countries:
    if city in countries[country]:
        print(f"INFO : {city} is a city in {country}")
        found = True
        break

if not found:
    print(f"ERROR : Country for {city} is not found")
```

```
sharkDistance = int(input("Distance du requin: "))
sharkSpeed = int(input("Vitesse du requin: "))
pontoonDistance = int(input("Distance jusqu'à l'île: "))
yourSpeed = int(input("Votre vitesse: "))
dolphin = input("Y a-t-il un dauphin (true/false)? ")

if dolphin == "true":
    sharkSpeed = sharkSpeed / 2

time_shark = sharkDistance / sharkSpeed
time_you = pontoonDistance / yourSpeed

if time_you < time_shark:
    print("Sauvé du requin !")
else:
    print("Le requin a mangé le nageur !")
```
